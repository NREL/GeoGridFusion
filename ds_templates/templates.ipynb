```python
import dask.array as da
import numpy as np
import xarray as xr

pvgis_template = xr.open_dataset("pvgis-sample-data.nc", engine="h5netcdf")

# or should we call this when required instead of saving
xr.Dataset(
    data_vars={
        name: (("gid", "time"), np.empty((0, len(pvgis_template.time)))) 
        for name in list(pvgis_template.data_vars)
    },
    coords={
        "time": pvgis_template.time.values,
        "gid": np.array([]),
    }
).to_netcdf("pvgis-tmy-template.nc", engine="h5netcdf")
```
