```python
import geogridstore
import numpy as np
from pyproj import CRS, Transformer
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
# Generate points with 2 km spacing
distance_km = 100
points = geogridstore.index.generate_spherical_grid_fixed(distance_km)

# Extract latitudes and longitudes for plotting
lats, lons = points[:, 0], points[:, 1]

# Plot using Cartopy
fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection=ccrs.Orthographic())
ax.stock_img()
ax.coastlines()

# Plot points
ax.scatter(lons, lats, color="red", s=1, transform=ccrs.PlateCarree())
ax.set_title(f"Uniform Spherical Grid with {distance_km} km Spacing (Fixed)")
plt.show()
```
